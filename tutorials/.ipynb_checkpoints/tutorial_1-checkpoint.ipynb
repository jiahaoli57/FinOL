## Generating Simulated Data
We will start by generating some random stochastic asset price data to use throughout the tutorial. This includes:

- Simulating daily OHLCV (OPEN, HIGH, LOW, CLOSE, and VOLUME) data for 1 assets over 100 days

```python
import pandas as pd
import numpy as np


def generate_ohlc(num_days):
    low_prices = np.random.uniform(100, 102, size=num_days)
    high_prices = np.random.uniform(low_prices, 105)

    # 生成开盘价（在最高价和最低价之间随机选择）
    open_prices = np.random.uniform(low_prices, high_prices)

    # 生成收盘价（在最高价和最低价之间随机选择）
    close_prices = np.random.uniform(low_prices, high_prices)

    # 生成成交量（范围为1000-5000之间）
    volume = np.random.randint(1000, 5000, size=num_days)

    return pd.DataFrame({
    'OPEN': open_prices,
    'HIGH': high_prices,
    'LOW': low_prices,
    'CLOSE': close_prices,
    'VOLUME': volume
    })


np.random.seed(0)
num_days = 100

df = generate_ohlc(num_days)

df['DATE'] = pd.date_range('2023-01-01', periods=num_days)
df = df.reindex(columns=['DATE', 'OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME'])
df
```

```python
import matplotlib.pyplot as plt

# 创建图形和轴对象
fig, ax1 = plt.subplots()

# 绘制'OPEN', 'HIGH', 'LOW', 'CLOSE'
ax1.plot(df['OPEN'], color='blue', label='OPEN')
# ax1.plot(df['HIGH'], color='green', label='HIGH')
# ax1.plot(df['LOW'], color='red', label='LOW')
# ax1.plot(df['CLOSE'], color='purple', label='CLOSE')

# 设置左边y轴标签和标题
ax1.set_ylabel('Price')
ax1.set_title('Stock Prices')

# 创建第二个y轴
ax2 = ax1.twinx()

# 绘制'VOLUME'
ax2.plot(df['VOLUME'], color='orange', label='VOLUME')

# 设置右边y轴标签
ax2.set_ylabel('Volume')

# 添加图例
lines = ax1.get_lines() + ax2.get_lines()
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

# 自动调整日期标签的格式
fig.autofmt_xdate()

# 显示图形
plt.show()
```
